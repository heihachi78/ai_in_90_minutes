## AI - Artificial Intelligence

Minden olyan szoftver megoldás, amely az emberi viselkedést, gondolkodást, cselekvést, érzékelést próbálja másolni, szimulálni.

Ebben az értelemben bizony ide tartozik az egymásba ágyazott if-ek (az emberi döntéshozó, problémamegoldó képességet próbálják szimulálni) is és ide tartoznak a nagy nyelvi modellek is (tanulás, érvelés, nyelvértés képessége) és a kettő között még nagyon-nagyon sok minden más is.

Példák az a_star és az if_as_ai könyvtárakban. Ezek az algoritmusok a döntéseiket egyszerű feltételek vagy heurisztika alapján hozzák meg, előzetes ismeretek, tanítás és memória nélkül.






### AI típusok

#### ANI - Artificial Narrow Inteligence

A fenti két példában látott AI képességei egy adott specifikus területre korlátozódnak: egy adott feladatban kiválóan teljesítenek, de más jellegű problémákat nem tudnak megoldani, a tudásukat nem tudják átvinni más területre. Az ilyen jellegű AI rendszereket gyenge vagy szűk AI-nak nevezzük. Jelenleg gyakorlatilag bármilyen AI-al találkozik is az ember, az ebbe a kategóriába tartozik, még akkor is, ha az a jelenlegi legokosabb nagy nyelvi modell, vagy mondjuk a Netflix ajánló rendszere, vagy az általam írt if-ekkel operáló "F1 pilóta".

#### AGI - Artificial General Intelligence

Egy AI rendszer akkor érné el az AGI szintet, bármilyen intellektuális feladatot el tudna végezni, amit egy ember meg tud csinálni. Ide még nem jutott el egy AI rendszer sem, de ha egyszer lesz ilyen, akkor például egy adott területen megszerzett ismereteit tudná alkalmazni egy másik területen.

#### ASI - Artificial Superintelligence

Ide elvileg azok az AI rendszerek tartoznának, amelyek minden területen meghaladják az emberi képességeket, de ez jelenleg elméleti szinten létezik csak, és nem nagyon fenyeget a veszélye, hogy hamarosan ilyet látni fogunk.

## ML - Machine Learning

A gépi tanulás az AI olyan részterülete, amely olyan algoritmusokra összpontosít, amelyek képesek tanulni és fejlődni anélkül, hogy explicit módon programoznák őket.

## Gépi tanulás altípusok



### 1. Felügyelt tanulás
- **Definíció**: Tanulás címkézett tanítóadatokból új, nem látott adatokon való előrejelzések készítése céljából
- **Folyamat**: Az algoritmus bemeneti-kimeneti párokból tanul
- **Típusok**:
  - **Osztályozás**: Diszkrét kategóriák vagy osztályok előrejelzése
    - Példák: Email spam-érzékelés, képfelismerés, orvosi diagnosztika
  - **Regresszió**: Folytonos numerikus értékek előrejelzése
    - Példák: Háziár-előrejelzés, részvényár-előrejelzés, időjárás-előrejelzés
- **Gyakori algoritmusok**: Lineáris regresszió, döntési fák, random forest, support vektor gépek

### 2. Felügyelet nélküli tanulás
- **Definíció**: Rejtett minták felismerése az adatokban címkézett példák nélkül
- **Folyamat**: Az algoritmus struktúrát fedez fel az adatokban célkimenetek nélkül
- **Típusok**:
  - **Klaszterezés**: Hasonló adatpontok csoportosítása
    - Példák: Vásárlói szegmentáció, génszekvenálás, piackutatás
  - **Asszociáció**: Különböző változók közötti kapcsolatok feltárása
    - Példák: Kosárelemzés, ajánlórendszerek
  - **Dimenziócsökkentés**: Adatok egyszerűsítése a fontos információk megőrzése mellett
    - Példák: Adatvizualizáció, jellemzőkiválasztás, tömörítés
- **Gyakori algoritmusok**: K-means klaszterezés, hierarchikus klaszterezés, PCA, DBSCAN

### 3. Megerősítéses tanulás
- **Definíció**: Tanulás a környezettel való interakció révén jutalmak és büntetések használatával
- **Folyamat**: Az ágens próbálkozás és tévedés útján tanulja meg a kumulatív jutalom maximalizálását
- **Főbb komponensek**:
  - **Ágens**: A tanuló vagy döntéshozó
  - **Környezet**: A világ, amelyben az ágens működik
  - **Cselekvések**: Az ágens számára elérhető választások
  - **Jutalmak**: Visszajelzés a környezettől
- **Példák**: Játékok (sakk, go), robotika, autonóm járművek, kereskedési algoritmusok
- **Gyakori algoritmusok**: Q-learning, policy gradient módszerek, actor-critic módszerek

### 4. Félig felügyelt tanulás
- **Definíció**: Felügyelt és felügyelet nélküli tanulás kombinációja címkézett és címkézetlen adatok használatával
- **Felhasználási esetek**: Amikor a címkézett adatok ritkák vagy költséges beszerzésűek
- **Példák**: Szövegklasszifikáció korlátozott címkézett dokumentumokkal, orvosi képalkotás kevés szakértői annotációval

## Mély tanulás és neurális hálózatok

A mély tanulás a gépi tanulás olyan specializált részterülete, amely több rétegű (ezért "mély") mesterséges neurális hálózatokat használ az adatok összetett mintáinak modellezésére és megértésére.

### Neurális hálózatok alapjai
- **Definíció**: Biológiai neurális hálózatok által inspirált számítási rendszerek
- **Struktúra**: Rétegekbe szervezett, összekapcsolt csomópontok (neuronok)
- **Komponensek**:
  - **Bemeneti réteg**: Adatokat fogad
  - **Rejtett rétegek**: Információkat dolgoz fel (több réteg = "mély")
  - **Kimeneti réteg**: Végső eredményeket produkál
  - **Súlyok és torzítások**: A hálózat viselkedését meghatározó paraméterek

### Neurális hálózat típusok

#### 1. Előrecsatolt neurális hálózatok
- **Struktúra**: Az információ egy irányban áramlik a bemenettől a kimenet felé
- **Felhasználási esetek**: Alapvető osztályozási és regressziós feladatok
- **Példa**: Házárak előrejelzése jellemzők alapján

#### 2. Konvolúciós neurális hálózatok (CNN-ek)
- **Specializáció**: Rács-szerű adatok (képek, videók) feldolgozása
- **Főbb jellemzők**: Konvolúciós rétegek, pooling rétegek, jellemzőtérképek
- **Alkalmazások**: Képfelismerés, orvosi képalkotás, autonóm járművek
- **Példa**: Objektumok azonosítása fényképeken

#### 3. Visszatérő neurális hálózatok (RNN-ek)
- **Specializáció**: Időbeli függőségekkel rendelkező szekvenciális adatok
- **Főbb jellemzők**: Memóriacellák, visszacsatolási hurkok
- **Változatok**: LSTM (Long Short-Term Memory), GRU (Gated Recurrent Unit)
- **Alkalmazások**: Természetes nyelvfeldolgozás, idősor-előrejelzés, beszédfelismerés
- **Példa**: Nyelvfordítás, chatbotok

#### 4. Transformer hálózatok
- **Innováció**: Figyelemmechanizmusok szekvenciák feldolgozásához
- **Főbb jellemzők**: Önfigyelem, párhuzamos feldolgozás
- **Alkalmazások**: Nagy nyelvi modellek (GPT, BERT), gépi fordítás
- **Példa**: Modern AI chatbotok és szöveggenerálás

### Mély tanulás alkalmazások
- **Számítógépes látás**: Kép- és videóelemzés
- **Természetes nyelvfeldolgozás**: Szövegmegértés és -generálás
- **Beszédfelismerés**: Beszéd szöveggé konvertálása
- **Játékok**: Összetett játékok elsajátítása (AlphaGo, sakk)
- **Gyógyszerkutatás**: Molekuláris elemzés és előrejelzés
- **Autonóm rendszerek**: Önvezető autók, drónok

## AI alkalmazások és fő célok

Az AI különböző területeken szolgál különféle célokat, átalakítva azt, ahogyan dolgozunk, kommunikálunk és problémákat oldunk meg.

### Az AI elsődleges céljai

#### 1. Automatizálás és hatékonyság
- **Cél**: Ismétlődő feladatok automatizálása és a működési hatékonyság növelése
- **Előnyök**: Csökkentett emberi hiba, 24/7 működés, költségmegtakarítás
- **Példák**: 
  - Gyártórobotok
  - Automatizált ügyfélszolgálati chatbotok
  - Adatbeviteli és feldolgozási rendszerek

#### 2. Döntéstámogatás és analitika
- **Cél**: Segítség az embereknek jobb, adatvezérelt döntések meghozatalában
- **Előnyök**: Gyorsabb elemzés, mintafelismerés, prediktív betekintések
- **Példák**:
  - Pénzügyi kockázatértékelés
  - Orvosi diagnózis-segítség
  - Üzleti intelligencia és előrejelzés

#### 3. Mintafelismerés és előrejelzés
- **Cél**: Összetett minták azonosítása az adatokban és jövőbeli eredmények előrejelzése
- **Előnyök**: Korai felismerés, trendelemzés, proaktív tervezés
- **Példák**:
  - Csalásfelismerés a bankszektorban
  - Prediktív karbantartás a gyártásban
  - Időjárás-előrejelzés és klímamodellezés

#### 4. Ember-számítógép interakció
- **Cél**: A technológia intuitívabbá és hozzáférhetőbbé tétele
- **Előnyök**: Természetes kommunikáció, jobb felhasználói élmény
- **Példák**:
  - Hangasszisztensek és okoshangszórók
  - Nyelvfordítási szolgáltatások
  - Gesztus- és arcfelismerő rendszerek

#### 5. Tartalomkészítés és fejlesztés
- **Cél**: Különböző típusú tartalmak generálása, módosítása és javítása
- **Előnyök**: Kreatív segítség, személyre szabás, minőségjavítás
- **Példák**:
  - AI-generált művészet és zene
  - Automatizált tartalomírás
  - Fénykép- és videófejlesztés

Függetlenül attól, hogy a technikai aspektusok, gyakorlati alkalmazások vagy társadalmi hatások érdeklik, az AI számos lehetőséget kínál a felfedezésre és hozzájárulásra.
